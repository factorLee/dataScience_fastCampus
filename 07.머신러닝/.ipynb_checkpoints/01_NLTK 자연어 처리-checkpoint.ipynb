## 2. 어간추출
    - NLTK의 PorterStemmer, LancasterStemmer 사용

```python
from nltk.stem import PorterStemmer, LancasterStemmer

st1 = PorterStemmer()
st2 = LancasterStemmer()

words = ["fly", "flies", "flying", "flew", "flown"]

print("Porter Stemmer :", [st1.stem(w) for w in words])
print("Lancater Stemmer :", [st2.stem(w) for w in words])
```

## 2.1 표제어 추출
    - NLTK의 WordNetLemmatizer 사용

```python
from nltk.stem import WordNetLemmatizer

lm = WordNetLemmatizer()

[lm.lemmatize(w, pos="v") for w in words]
```

```python
# pos_tag 사용 (품사 붙여줌)

from nltk.tag import pos_tag

sentence = "Emma refused to permit us to obtain the refuse permit"
tagged_list = pos_tag(word_tokenize(sentence))
tagged_list

# 튜플의 리스트로 결과가 나옴
```

```python
# 토큰
def tokenizer(doc):
    return ["/".join(p) for p in tagged_list]

tokenizer(sentence)
```

## NLTK 빈도분석

```python
from nltk import Text

text = Text(word_tokenize(emma_raw))
text
```

```python
import matplotlib.pyplot as plt

# 단어 빈도
text.plot(20) # 가장많이 사용되는 20개의 단어 표시
plt.show()
```

```python
# 단어 위치
text.dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])
```

```python
# 문맥 인쇄
text.concordance("Emma")
```

```python
# 유사단어 검색
text.similar("Emma", 1)
```

```python
# 공통 문맥 인쇄
text.common_contexts(["Emma", "she"], 1)
```

```python
fd = text.vocab()
type(fd)
```

```python
from nltk import FreqDist

stopwords = ["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear"]
emma_tokens = pos_tag(word_tokenize(emma_raw))
names_list = [t[0] for t in emma_tokens if t[1] == "NNP" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
fd_names.N(), fd_names["Emma"], fd_names.freq("Emma") # 전체 이름중에 12프로정도가 emma로 나타남
```

```python
fd_names.most_common(5)
```

```python
fd_names.plot(20)
plt.show()
```

```python
!pip install wordcloud
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis("off")
plt.show()
```

```python
bible_raw = nltk.corpus.gutenberg.raw('bible-kjv.txt')
print(bible_raw[:702])
```

```python
from nltk import FreqDist

stopwords = ["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear"]
shakespeare_tokens = pos_tag(word_tokenize(bible_raw))
names_list = [t[0] for t in shakespeare_tokens if t[1] == "NNP" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis("off")
plt.show()
```
