```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
rv = sp.stats.norm(loc=1, scale=2)
```

```python
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('확률밀도함수')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

```python
# 연습문제 8.1.1
rv_2 = sp.stats.norm(loc=0, scale=0.1)
```

```python
xx = np.linspace(-1, 1, 100)
pdf = rv_2.pdf(xx)
plt.plot(xx, pdf)
plt.title('확률밀도함수')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```
