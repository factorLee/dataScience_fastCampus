```python
import seaborn as sns
import matplotlib.pylab as plt
import statsmodels.api as sm
```

```python
law_data = pd.read_csv('fitts_law04.csv')

dfX0 = pd.DataFrame(law_data["ID"])
dfX = sm.add_constant(dfX0)
dfy = pd.DataFrame(law_data["MT"])
```

```python
model_fitts = sm.OLS(dfy, dfX)
result_fitts = model_fitts.fit()
print(result_fitts.summary())
```

```python
sns.pairplot(law_data[["MT", "ID"]])
plt.show()
```

```python
sns.lmplot(x="ID", y="MT", data=law_data, line_kws={'color':"red"})
plt.xlabel("ID")
plt.ylabel("MT")
plt.title("Fitt's law regression")
plt.show()
```
